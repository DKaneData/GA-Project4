```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv('../Data/final_inflation_adj_data.csv')
df.head()
```

```python
yearly_sums = df.groupby('year').sum('budget')
yearly_sums
```

```python
temp_df = df.groupby('year').nunique(['movies'])
temp_df

num_movies_per_year = temp_df.filter(['name'])
num_movies_per_year.rename({'name' : 'num_movies' }, axis = 1, inplace = True)
num_movies_per_year
```

```python
yearly_sums = yearly_sums.merge(num_movies_per_year, left_on = yearly_sums.index, right_on = num_movies_per_year.index)
yearly_sums
```

```python
yearly_sums.rename({'key_0' : 'year'}, axis = 1, inplace = True)
```

#### Number of Movies Per Year caps out at 200. The repository we pulled it from says they scraped 220 movies per year, and we dropped some movies where we were missing data. It may not be all movies that year.

Given that there's no more than 200 movies per year, this may not be all movies released that year - maybe just the top movies selected based on some factor (number of votes, budget/gross, etc.)

```python
yearly_sums
```

```python
years = yearly_sums['year']
num_movies = yearly_sums['num_movies']
profit_adj = yearly_sums['profit_adj']

fig, ax1 = plt.subplots(figsize = (10, 10))
ax2 = ax1.twinx()

ax1.bar(years, num_movies, color = 'blue', alpha = 0.5, label = 'Number of Movies')
ax1.set_xlabel('Year', fontsize = 15, color = 'black')
ax1.set_ylabel('Number of Movies', color = 'blue', fontsize = 15)
ax1.tick_params(axis = 'x', labelcolor = 'black') # Make the X Axis black
ax1.tick_params(axis = 'y', labelcolor = 'blue') # Make the Y axis blue to match the color

ax2.plot(years, profit_adj / 1_000_000_000, color = 'black', marker = '*', linewidth = 2, label = 'Total Industry Profit (Adjusted)')
ax2.set_ylabel('Total Industry Profit in Billions of $\n(Adjusted for Inflation)', color = 'black', fontsize = 15)

#Added a line for 2007 because that is when Netflix started it's streaming service 
plt.axvline(2007, linewidth = 3, color = 'red')

plt.title('Number of Movies and Total Industry Profit (Inflation Adjusted)\nby Year', fontsize = 20)

plt.savefig('../Images/num_movies_and_total_industry_profit_by_year.png', dpi = 192)

plt.show();
```

```python
# Looking at the Mean Budget / Gorss / Profit by Year
yearly_means = df.groupby('year').mean('budget_in_millions').reset_index()
yearly_means
```

```python
yearly_medians = df.groupby('year').median('budget_in_millions').reset_index()
yearly_medians
```

```python
plt.plot(yearly_medians['year'],
         yearly_medians['gross'] / 1_000_000,
         label = 'Unadjusted', color = 'orange')

plt.plot(yearly_medians['year'],
         yearly_medians['gross_adj'] / 1_000_000,
         label = 'Inflation Adjusted', color = 'green')

plt.xlabel('Year', fontsize = 15)
plt.xticks(fontsize = 10)
plt.ylabel('Median Movie Gross\nin Millions of $', fontsize = 15)
plt.yticks(fontsize = 10)

plt.axvline(2007, linewidth = 3, color = 'blue')

plt.title('Median Movie Gross by Year', fontsize = 20)

plt.tight_layout()

plt.legend()

plt.savefig('../Images/median_movie_gross_by_year.png', dpi = 192);
```

```python
plt.plot(yearly_means['year'],
         yearly_means['gross'] / 1_000_000,
         label = 'Unadjusted', color = 'orange')

plt.plot(yearly_means['year'],
         yearly_means['gross_adj'] / 1_000_000,
         label = 'Inflation Adjusted', color = 'green')

plt.xlabel('Year', fontsize = 15)
plt.xticks(fontsize = 10)
plt.ylabel('Mean Movie Gross\nin Millions of $', fontsize = 15)
plt.yticks(fontsize = 10)

plt.axvline(2007, linewidth = 3, color = 'blue')

plt.title('Mean Movie Gross by Year', fontsize = 20)

plt.tight_layout()

plt.legend()

plt.savefig('../Images/mean_movie_gross_by_year.png', dpi = 192);
```

```python
plt.plot(yearly_medians['year'],
         yearly_medians['budget_adj'] / 1_000_000,
         label = 'Median Budget', color = 'red', marker = '*')

plt.plot(yearly_medians['year'],
         yearly_medians['gross_adj'] / 1_000_000,
         label = 'Median Gross', color = 'olive', marker = 'o')

plt.plot(yearly_medians['year'],
         yearly_medians['profit_adj'] / 1_000_000,
         label = 'Median Profit', color = 'purple', marker = '.')

plt.axvline(2007, linewidth = 3, color = 'blue')

plt.axhline(0, linewidth = 3, color = 'black')

plt.ylabel('Millions of Dollars', fontsize = 15)
plt.yticks(fontsize = 10)
plt.xlabel('Year', fontsize = 15)
plt.xticks(fontsize = 10)

plt.title('Median Movie Costs & Revenue Per Year\n(Inflation Adjusted)', fontsize = 20)

plt.legend()

plt.tight_layout()

plt.savefig('../Images/median_movie_costs_revenue_by_year.png', dpi = 192);
```

```python
df['is_profitable'] = np.where(df['profit_adj'] > 0, 1, 0)
df.tail()
```

```python
df.head()
```

```python
yearly_sums = df.groupby('year').sum('is_profitable')
profitable_films_by_year = pd.DataFrame(yearly_sums['is_profitable'])
profitable_films_by_year
```

```python
num_movies_per_year[num_movies_per_year.index != 2020]
```

```python
profitable_movies_by_year = profitable_films_by_year.merge(num_movies_per_year[num_movies_per_year.index != 2020], how = 'inner', left_on = profitable_films_by_year.index,
                               right_on = num_movies_per_year[num_movies_per_year.index != 2020].index)
profitable_movies_by_year
```

```python
profitable_movies_by_year['%_profitable'] = profitable_movies_by_year['is_profitable'] / profitable_movies_by_year['num_movies']
profitable_movies_by_year
```

```python
profitable_movies_by_year.rename({'key_0' : 'year'}, axis = 1, inplace = True)
```

```python
profitable_movies_by_year
```

```python
plt.plot(profitable_movies_by_year['year'],
         profitable_movies_by_year['%_profitable'],
         color = 'red', marker = '*')

plt.axvline(2007, linewidth = 3, color = 'blue')
plt.axhline(.5, linewidth = 3, color = 'black')

plt.ylabel('% Profitable', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)

plt.title('% of Movies that are Profitable per Year', fontsize = 20)

plt.tight_layout()

plt.savefig('../Images/percent_movies_profitable_per_year.png', dpi = 192);
```
