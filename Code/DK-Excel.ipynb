```python
# Imported the Imputed Data set since it has the budget filled all movies
df = pd.read_csv('./Data/imputed.csv')
df.head()
```

```python
# Create function where pass in the year and the month, and then it returns from the bls_dict that month and year
def get_inflation_adj_val(year, month):
    return bls_dict[month][year]
```

```python
df['inf_adjmnt_value'] = df.apply(lambda x: get_inflation_adj_val(x['year'], x['month']), axis=1)
    # Take the data frame, then x is each row and each row apply the get_inflation_adj_val function.
    # Get the year and month from each row, put it into the get_inflation_adj_val function, then it spits out the BLS CPI for that year and month.
```

```python
# The inflation adjustment value / divided by 100 to turn it into a percentage - so 258 is 2.58 times the baseline and 82.7 is 0.827 the baseline
# Then divide the budget by that - if inflation adjustment fraction is 1, then dividing by one. But if inflation adjustment fraction is 2.58, dividing by 2.58 to make it the baseline for 1983
df['budget_cpi_adj'] = df['budget'] / (df['inf_adjmnt_value']/100)
df
```

```python
df['gross_cpi_adj'] = df['gross'] / ( df['inf_adjmnt_value']/100)
df
```

```python
df['profit_cpi_adj'] = df['gross_cpi_adj'] - df['budget_cpi_adj']
df.head()
```

```python
df['budget'] = df['budget'] / 1_000_000
df['gross'] = df['gross'] / 1_000_000
df['budget_cpi_adj'] = df['budget_cpi_adj'] / 1_000_000
df['gross_cpi_adj'] = df['gross_cpi_adj'] / 1_000_000
df['profit_cpi_adj'] = df['profit_cpi_adj'] / 1_000_000

df
```

```python
df.columns
```

```python
cols_to_rename = { 'budget' : 'budget_in_millions',
                  'gross' : 'gross_in_millions',
                  'budget_cpi_adj' : 'budget_cpi_adj_in_millions',
                  'gross_cpi_adj' : 'gross_cpi_adj_in_millions',
                  'profit_cpi_adj' : 'profit_cpi_adj_in_millions'
}

df.rename(mapper = cols_to_rename, axis = 1, inplace = True)

df.columns
```

```python
df
```

```python
temp_df = df.groupby('year').nunique(['movies'])
temp_df
```

```python
# Got the number of movies per year
num_movies_per_year = temp_df.filter(['name'])
num_movies_per_year.rename({'name' : 'num_movies' }, axis = 1, inplace = True)
num_movies_per_year
```

```python
yearly_means = df.groupby('year').mean('budget_in_millions').reset_index()
yearly_means
```

```python
# Merging in # of movies per year
yearly_means = yearly_means.merge(num_movies_per_year, left_on = yearly_means['year'], right_on = num_movies_per_year.index)
yearly_means.drop(columns = 'key_0', inplace = True)
yearly_means.tail()
```

```python
yearly_medians = df.groupby('year').median('budget_in_millions').reset_index()
yearly_medians
```

```python
yearly_medians = yearly_medians.merge(num_movies_per_year, left_on = yearly_medians['year'], right_on = num_movies_per_year.index)
```
